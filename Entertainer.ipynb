```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Entertainer - Basic Info.xlsx')
df1=pd.read_excel('Entertainer - Breakthrough Info.xlsx')
df2=pd.read_excel('Entertainer - Last work Info.xlsx')
```

```python
df.head(2)
```

```python
df5 = df1.merge(df, how='inner', on='Entertainer')          
df5.head(2)
```

```python
df6 = df5.merge(df2, how='inner', on='Entertainer')         
df6.head(2)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df6.corr(), annot=True)
plt.show()
```

```python
counts = df6['Gender (traditional)'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(counts, labels=counts.index, shadow=True, startangle=45, autopct='%1.1f%%')
plt.title('Gender (traditional)')
plt.legend()
plt.show()
```
